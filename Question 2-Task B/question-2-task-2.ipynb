```python
# Import necessary libraries
import os
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

# Define the audio files for each genre
audio_files = {
    "Pop": "/kaggle/input/audiofiles/Songwriterz_Alive_POP.mp3",  
    "HipHop": "/kaggle/input/audiofiles/Tom_Orlando_-_Into_You_HIPHOP.mp3",  
    "Classical": "/kaggle/input/audiofiles/Shaunas_Theme_Jorgen_Kruse_CLASSICAL.mp3", 
    "Jazz": "/kaggle/input/audiofiles/Athens_Pierce_Murphy_JAZZ.mp3" 
}

# Load audio files and generate spectrograms
for genre, file_path in audio_files.items():
    # Load the audio file
    audio, sr = librosa.load(file_path, sr=None)
    
    # Generate the spectrogram using STFT
    stft = librosa.stft(audio)
    spectrogram = np.abs(stft)
    
    # Plot the spectrogram in a separate figure
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(librosa.amplitude_to_db(spectrogram, ref=np.max),
                             sr=sr, x_axis="time", y_axis="log")
    plt.colorbar(format="%+2.0f dB")
    plt.title(f"Spectrogram of {genre} Genre")
    plt.show()
```
